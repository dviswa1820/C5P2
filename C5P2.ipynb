```python
#creating an additional column to discretize age. For clarity purposes here the bins are categorical text. 
#They will label or one hot encoded when building the model
def create_person(credit):
    for i in range(0,len(credit)):
        if(credit.loc[i,'AGE']<35):
            credit.loc[i,'person']='young'
        elif(credit.loc[i,'AGE']>=35 and credit.loc[i,'AGE']<50): 
            credit.loc[i,'person']='middle_aged'
        else:
            credit.loc[i,'person']='old'

create_person(credit) 
```
